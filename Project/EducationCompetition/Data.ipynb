## 결측값 확인

```python
# 각 칼럼의 결측값 개수 계산
missing_values = df.isnull().sum()

# 결측값 비율(%)도 함께 출력
missing_percentage = (missing_values / len(df)) * 100
missing_info = pd.DataFrame({"결측값 개수": missing_values, "결측값 비율(%)": missing_percentage})

# (선택) 결측값 개수가 있는 칼럼만 출력
print(missing_info[missing_info["결측값 개수"] > 0])
```

## 칼럼명 수정

```python
df.columns = [
    col.replace('B1', 'A3') if '문B1' in col and '귀하께서 작년에 참여한 학습경험이 전반적으로' in col else col
    for col in df.columns
]
```

```python
# A3가 포함된 칼럼명만 출력
a3_columns = [col for col in df.columns if 'A3' in col]
print(a3_columns)
```

## 종속 변수 설정

```python
# '문A1-1)'와 '문B1'이 포함된 컬럼에서 1이 하나라도 있으면 1, 아니면 0
df['참여여부'] = ((df.filter(like='문A1-1)').eq(1).any(axis=1)) | (df.filter(like='문B1').eq(1).any(axis=1)))

# 0과 1로 표시되도록 하기 위해 int형으로 변환
df['참여여부'] = df['참여여부'].astype(int)
```
